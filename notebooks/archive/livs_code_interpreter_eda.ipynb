```python
import re

re.findall(
r'-?[\d]*[.][\d]+',
"AI:  The estimated average treatment effect of 'receive' on 'outcome' using instrumental variable analysis with two-stage least squares is approximately -0.022."
)
```

```python
async def flu_iv_prompt(prompt, model="gpt-3.5-turbo"):
    async with CodeInterpreterSession(model=model) as session:
        # define the user request
        user_request = prompt
        files = [
            File.from_path("/home/tliu/learning-ivs/data/flu_clean.dta"),
        ]

        # generate the response
        response = await session.generate_response(
            user_request, files=files
        )

        # output to the user
        print("AI: ", response.content)
        #for file in response.files:
        #    file.show_image()
        
        # extract treatment effect estimate
        treat_effect = re.findall(r'-?[\d]*[.][\d]+', response.content)
        return float(treat_effect[0])
            
```
